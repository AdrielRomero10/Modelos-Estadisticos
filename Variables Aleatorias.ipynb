## Variable aleatoria uniforme discreta

Decimos que una variable aleatoria $X$ tiene distribución **uniforme discreta** en el conjunto $\{x_1, x_2, \dots, x_n\}$ si su función de densidad está dada por
$$f_{X}(x) = \begin{cases} \frac{1}{n}, \quad \text{si } x\in \{x_1, x_2, \dots, x_n\} \\ 0, \quad \text{e.o.c} \end{cases}$$

Se llama *uniforme* porque cada uno de sus posibles resultados de $X$ tienen la misma probabilidad.

**Notación:** $X \sim Unif(x_1, x_2, \dots, x_n)$.

La esperanza y la varianza de $X$ están dadas por:

$$\mathbb{E}(X) = \sum_{x}xf_{X}(x) = \frac{x_n + x_1}{2}$$
y
$$\text{Var}(X) = \mathbb{E}[(X-\mathbb{E}(x))^2] = \frac{(x_n - x_1 + 1)^2 -1}{12}$$

$\color{purple}{\text{Ejercicio 1.}}$ Verifica que, en efecto, las formulas anteriores son ciertas.

**Ejemplo** Se lleva a cabo una rifa donde los boletos están enumerados del $00$ al $99$. Si $Y$ es la variable aleatoria definida como el número del boleto ganador, entonces:  
$$\mathbb{P}(Y=k)=\begin{cases}
  \frac{1}{100} &\quad \text{si } k=00,01,\dots,99 \\
  0 &\quad \text{en otro caso}
\end{cases}$$
Consideremos que el premio de la rifa se determina a partir del número premiado de la siguiente forma: $X=Y+1$, donde $X$ es el monto del premio en pesos y $Y$ es el número premiado, entonces $X$ es una variable aleatoria, pues es una función de $Y$, y además se tiene
  $$f_{X}(k)=\begin{cases}
    \frac{1}{100} &\quad \text{si } x=1,2,\dots,100 \\
    0 &\quad \text{en otro caso}
  \end{cases}$$
  
$\color{purple}{\text{Ejercicio 2.}}$ Calcular su esperanza y varianza.

Supongamos que nos interesa calcular la probabilidad de que el premio sea mayor a $80. Entonces:

$$\mathbb{P}(X>80) = \sum_{x=81}^{100}\frac{1}{100}=20 \frac{1}{100}=0.2 $$

```python
# Distribución uniforme discreta

# Vamos a definir el límite inferior y superior de la uniforme
lim_inf = 0
lim_sup = 10

# Creamos los valores posibles dentro del rango de la variable aleatoria
# np.arange crea un arreglo con valores desde lim_inf hasta lim_sup-1
x = np.arange(lim_inf, lim_sup)

# Calculamos la función de densidad de la uniforme
pmf = np.full_like(x, 1/(lim_sup-lim_inf), dtype = float) # crea un arreglo del mismo tamaño que x, donde todos sus elementos tienen el mismo valor

# Graficamos la función de densidad
plt.figure(figsize=(10,5)) # tamaño de la figura
plt.bar(x, pmf , width = 0.2, color='skyblue', edgecolor='red') # gráfico de barras
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Función de densidad de una variable aleatoria uniforme discreta')
plt.grid(True)
plt.show()
```

## Variable aleatoria Bernoulli con parámetro $p\in(0,1)$

Este modelo se aplica a un experiemento cuyo espacio muestral tiene dos resultados
$$ \Omega = \{\text{éxito} , \text{fracaso}\}$$
y definimos
- $X(\{\text{éxito}\}) =1$
- $X(\{\text{fracaso}\}) =0$

Las probabilidades asociadas a este modelo son
- $\mathbb{P}(\{\text{éxito}\}) = p$
- $\mathbb{P}(\{\text{fracaso}\}) = 1-p$, 

donde $0<p<1$.

La función de densidad, está definida de la siguiente manera
$$ f_X(x) = \mathbb{P}(X=x) = \begin{cases} p^x(1-p)^{1-x}, \quad \text{si } x\in \{0,1\} \\
0, \quad \text{e.o.c} \end{cases}$$

**Notación:** $X\sim Ber(p)$

Mediante la combinación de v.a. Bernoulli es posible construir otras v.a.

La esperanza y varianza de $X\sim Ber(p)$, está dada por:

$$ \mathbb{E}[X] = \sum_{x=0}^{1} xf_{X}(x) = 0\cdot(1-p) + 1\cdot p = p$$
y
$$ \mathbb{E}[X^2] = \sum_{x=0}^{1} x^2 f_{X}(x) = 0^2\cdot(1-p) + 1^2\cdot p = p$$
entonces
$$\text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = p - p^2 = p(1-p) $$

## Variable aleatoria Binomial con parámetros $n\in\mathbb{Z}_{+}$ y $p\in(0,1)$

Es una distribución de probabilidad discreta que sirve para calcular la probabilidad de que ocurra un determinado número de eventos raros durante un intervalo dado (puede ser tiempo, lingitud, área, volumén, etc).

Esta v.a. tomavalores sobre el conjunto $\{0,1,2,...\}$ y tiene un parámetro $\lambda>0$, el cual representa el número de veces que se **espera** que ocurra un evento durante un intervalo dado.

Su función de densidad, está dado como sigue:
$$ f_X(x) = \mathbb{P}(X=x) = \begin{cases} e^{-\lambda}\frac{\lambda^x}{x!}, \quad \text{si } x\in \{0,1,...\} \\
0, \quad \text{e.o.c} \end{cases}$$

**Notación:** $X\sim \text{Poi}(\lambda)$.

Notemos que $\mathbb{E}[X]=\lambda$ y que $\text{Var}(X)=\lambda$.

**Ejemplo** Supongamos que el número de accidentes que ocurre en un punto en un día tiene distribución Poisson con parámetro $\lambda=2$,
- ¿cuál es la probabilidad de que en un día ocurran más de dos accidentes?
  $$ \begin{align*} \mathbb{P}(X>2) &= 1- \mathbb{P}(X\le 2) \\
    &= 1- [\mathbb{P}(X=0) + \mathbb{P}(X=1) + \mathbb{P}(X=2)] \\
    &= 1- \left[ e^{-2}\frac{2^0}{0!} + e^{-2}\frac{2^1}{1!} + e^{-2}\frac{2^2}{2!} \right] \\
    &= 1 - e^{-2}[1+2+2] = 1-5e^{-2} = 0.3233
    \end{align*} $$
- ¿cuál es la probabilidad de que ocurran más de dos accidentes sabiendo que por lo menos ocurre uno?
  $$ \begin{align*} \mathbb{P}(X>2 \mid X \ge 1) &= \frac{\mathbb{P}(\{X>2\} \cap \{X \ge 1\} )}{\mathbb{P}\{X \ge 1)\}} \\
       &= \frac{\mathbb{P}(\{X>2\})}{\mathbb{P}\{X \ge 1)\}} \\
       & = \frac{.3233}{1-\mathbb{P}(X<1)} = \frac{0.3233}{1-e^{-2}} = \frac{0.3233}{0.8646} = 0.3739
       \end{align*} $$

$\color{purple}{\text{Ejercicio 3.}}$ Usando el atributo `.cdf` [Poisson](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html) calcula las probabilidades anteriores. Y también graficar la función de densidad.

```python
# Solucion ejercicio 3

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetro de la distribución
lam = 2  

# --- (a) Probabilidad de que ocurran más de 2 accidentes ---
P_mas_de_2 = 1 - poisson.cdf(2, mu=lam)

# --- (b) Probabilidad de que ocurran más de 2 accidentes dado que ocurre al menos 1 ---
P_al_menos_1 = 1 - poisson.cdf(0, mu=lam)
P_mas_de_2_dado_1 = P_mas_de_2 / P_al_menos_1

# Mostrar resultados
print(f"P(X > 2) = {P_mas_de_2:.4f}")
print(f"P(X > 2 | X >= 1) = {P_mas_de_2_dado_1:.4f}")

# --- Graficar la función de masa de probabilidad (pmf) ---
x = np.arange(0, 10)  # valores de 0 a 9 accidentes
pmf = poisson.pmf(x, mu=lam)

plt.stem(x, pmf, basefmt=" ")  # <- corregido
plt.title("Distribución de Poisson con λ=2")
plt.xlabel("Número de accidentes (x)")
plt.ylabel("P(X = x)")
plt.show()

```

## ejemplo 

El número de éxitos en secuencia fija de n ensayos independientes de Bernoulli, con la misma probabilidad de éxito $(p)$

$\textcolor{cyan}{Ejemplo:}$ El número de artículos defectuosos en una muestra de 10 productos, si la probabilidad de ser defectuoso es $0.05$

Variable Ateatoria $(X)$

$X=$ Número de artículos defectuosos en la muestra de 10

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt


# 1. Definir los parámetros de la distribución

n = 10   # Número de ensayos
p = 0.3  # Probabilidad de éxito
k = 4    # Valor específico a evaluar


#  Crear la distribución binomial
X_binomial = stats.binom(n=n, p=p)

#  Calcular probabilidades

valores = np.arange(0, n + 1)
probabilidades = X_binomial.pmf(valores)


#  Mostrar información numérica

print("Distribución Binomial (n=10, p=0.3):")
print(f"P(X={k}) (4 éxitos): {X_binomial.pmf(k):.4f}")
print(f"Media: {X_binomial.mean():.2f}")
print(f"Varianza: {X_binomial.var():.2f}")


# 5. Crear la gráfica

plt.bar(valores, probabilidades, color='skyblue', edgecolor='black', width=0.6)
plt.title('Distribución Binomial (n=10, p=0.3)')
plt.xlabel('Número de éxitos (k)')
plt.ylabel('Probabilidad P(X=k)')
plt.xticks(valores)

# Resaltar el valor k=4
plt.bar(k, X_binomial.pmf(k), color='orange', edgecolor='black', width=0.6, label=f'P(X={k})={X_binomial.pmf(k):.3f}')

# Agregar etiquetas encima de las barras
for x, p_val in zip(valores, probabilidades):
    plt.text(x, p_val + 0.005, f'{p_val:.2f}', ha='center')

plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```


## Poisson


El número de eventos que ocurren en un intervalo fijo de tiempo o espacio, con una tasa promedio $(\lambda)$ conocida y constante.

$\textcolor{cyan}{Ejemplo:}$ El número de llamadas telefónicas que llegan a un call center en un periodo de una hora.

Variale Aleatoria $(X)$

$X=$ Número de llamadas recibidas en una hora

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt


# 1. Parámetros de la distribución Poisson

lam = 2.0  # Tasa promedio (λ)
k = 7      # Número de eventos a evaluar


# Crearmos la distribución Poisson

X_poisson = stats.poisson(mu=lam)


#  Calcularmos probabilidades

valores = np.arange(0, 15)  # Posibles valores de X (0 a 14)
probabilidades = X_poisson.pmf(valores)


#  Mostramos la información

print(f"Distribución Poisson (λ={lam}):")
print(f"P(X={k}) ({k} eventos): {X_poisson.pmf(k):.4f}")
print(f"Media (λ): {X_poisson.mean():.2f}")
print(f"Varianza (λ): {X_poisson.var():.2f}")


#  Graficamos la distribución

plt.bar(valores, probabilidades, color='skyblue', edgecolor='black', width=0.6)
plt.title(f'Distribución Poisson (λ={lam})')
plt.xlabel('Número de eventos (k)')
plt.ylabel('Probabilidad P(X=k)')
plt.xticks(valores)

# Resaltar el valor k=7
plt.bar(k, X_poisson.pmf(k), color='green', edgecolor='black', width=0.6, label=f'P(X={k})={X_poisson.pmf(k):.3f}')

# Agregar etiquetas sobre las barras
for x, p_val in zip(valores, probabilidades):
    plt.text(x, p_val + 0.005, f'{p_val:.2f}', ha='center')

plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

# Geométrica

## Descripción: 

El número de fracasos antes de obtener el primer éxito en ensayos de Bernoulli independientes.


$\textcolor{Cyan}{Ejemplo:}$ El número de lanzamientos fallidos de un tiro libre que realiza un jugador de baloncesto antes de conseguir su primer enceste.

Variable Aleatoria $(X)$

$X=$ Número de fallos antes del primer acierto

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt


#  Parámetros de la distribución geométrica

p = 0.3  # Probabilidad de éxito
k = 3    # Ensayo donde ocurre el primer éxito


# Crearmos  la distribución Geométrica

X_geometrica = stats.geom(p=p)


# Calcularmos las probabilidades

valores = np.arange(1, 11)  # Ensayos de 1 a 10
probabilidades = X_geometrica.pmf(valores)


# Mostramos   información numérica

print(f"Distribución Geométrica (p={p}):")
print(f"P(X={k}) → Primer éxito en el {k}° ensayo: {X_geometrica.pmf(k):.4f}")
print(f"Media (1/p): {X_geometrica.mean():.2f}")
print(f"Varianza ((1-p)/p²): {X_geometrica.var():.2f}")


#  Graficar la distribución

plt.bar(valores, probabilidades, color='green', edgecolor='black', width=0.6)
plt.title(f'Distribución Geométrica (p={p})')
plt.xlabel('Ensayo (k)')
plt.ylabel('Probabilidad P(X=k)')
plt.xticks(valores)

# Resaltar el valor k específico
plt.bar(k, X_geometrica.pmf(k), color='orange', edgecolor='green', width=0.6, label=f'P(X={k})={X_geometrica.pmf(k):.3f}')

# Agregar etiquetas encima de las barras
for x, p_val in zip(valores, probabilidades):
    plt.text(x, p_val + 0.005, f'{p_val:.2f}', ha='center')

plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

# Binomial Negativa (BN)

## Descripción

El número de fracasos antes de obtener r éxitos en ensayos de Bernoulli independientes

$\textcolor{cyan}{Ejemplo:}$ El número de lanzamientos fallidos de un dado antes de obtener un 6 por tercera vez $(r=3)$

Variable Aleatoria $(X)$

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

# 1. Parámetros de la distribución

r = 5    # Número de éxitos requeridos
p = 0.3  # Probabilidad de éxito
k = 10   # Número de fallos antes del r-ésimo éxito


#  Creamos la distribución Binomial Negativa

X_bn = stats.nbinom(n=r, p=p)


# Calculamos probabilidades

valores = np.arange(0, 25)  # Posibles números de fallos (0 a 24)
probabilidades = X_bn.pmf(valores)


# 4. Mostrar información numérica

print(f"Distribución Binomial Negativa (r={r}, p={p}):")
print(f"P(X={k}) ({k} fallos antes del {r}° éxito): {X_bn.pmf(k):.4f}")
print(f"Media (r*(1-p)/p): {X_bn.mean():.2f}")
print(f"Varianza (r*(1-p)/p²): {X_bn.var():.2f}")


#Graficamos la distribución

plt.bar(valores, probabilidades, color='skyblue', edgecolor='black', width=0.6)
plt.title(f'Distribución Binomial Negativa (r={r}, p={p})')
plt.xlabel('Número de fallos (k)')
plt.ylabel('Probabilidad P(X=k)')
plt.xticks(valores)

# Resaltar el valor k específico
plt.bar(k, X_bn.pmf(k), color='red', edgecolor='black', width=0.6,
        label=f'P(X={k})={X_bn.pmf(k):.3f}')

# Agregar etiquetas encima de las barras
for x, p_val in zip(valores, probabilidades):
    plt.text(x, p_val + 0.002, f'{p_val:.2f}', ha='center', fontsize=8)

plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

# Hipergeométrica

## Descripción

El número de éxitos en una muestra de tamaño n extraida sin reemplazo de una población finita.

$\textcolor{cyan}{Ejemplo:}$ En una caja con 20 bombillas, de las cuales 5 son defectuosas, se extrae una muestra de 4 bombillas sin devolverlas. $X$ es el número de bombillas defectuosas en la muestra

Variable Aleatoria $(X)$

$X=$ Número de bombillas defectuosas en la muestra de 4

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Parámetros del ejemplo
N_total = 100   # Tamaño total de la población (M)
N_exitos = 30   # Número de éxitos en la población (n)
n_muestra = 10  # Tamaño de la muestra (N)
k = 4           # Número de éxitos en la muestra

# Crear la distribución hipergeométrica
X_hiper = hypergeom(M=N_total, n=N_exitos, N=n_muestra)

# Valores posibles de éxitos en la muestra
x = np.arange(0, n_muestra + 1)
pmf = X_hiper.pmf(x)

# Mostrar resultados numéricos
print(f"Hipergeométrica (N_total={N_total}, N_exitos={N_exitos}, n_muestra={n_muestra}):")
print(f"P(X={k}) ({k} éxitos en la muestra): {X_hiper.pmf(k):.4f}")
print(f"Media: {X_hiper.mean():.2f}")
print(f"Varianza: {X_hiper.var():.2f}")

# Gráfica tipo stem (sin el argumento eliminado)
plt.figure(figsize=(8, 5))
plt.stem(x, pmf, basefmt=" ")  # ✅ quitamos 'use_line_collection=True'
plt.title(f"Distribución Hipergeométrica\nN={N_total}, éxitos={N_exitos}, muestra={n_muestra}")
plt.xlabel("Número de éxitos en la muestra (k)")
plt.ylabel("Probabilidad P(X=k)")
plt.grid(alpha=0.4)

# Resaltar P(X=k)
plt.stem([k], [X_hiper.pmf(k)], linefmt='r', markerfmt='ro', basefmt=" ", label=f"P(X={k})={X_hiper.pmf(k):.3f}")
plt.legend()
plt.show()
```

## Métodos de generación de números aleatorios rectangulares

La generación de variables aleatorias con esta distribución es importante porque las variables que tengan una distribución diferente, tendrán que usar a ésta como base.

- Cada posible resultado entre $a$ y $b$ tiene la misma probabilidad $1/n$.

Las variables generadas deben cumplir con:
- Los valores generados deben ser independientes y estar idénticamente distribuidos
- La secuencia generada debe ser lo más larga posible y ser reproducibles
- Debe permitir generar múltiples secuencias
- Que usen poca memoria

Históricamente se han usado cuatro tipos de métodos para generar sucesiones de números rectangulares:
- Métodos manuales
- Tablas de biblioteca
- Computación analógica
- Computación digital

### Generación pseudoaleatoria

La generación de los números aleatorios rectangulares debe realizarse a través de relaciones matemáticas de recurrencia. Por esta razón se consideran **pseudoaleatorios**, ya que el proceso para generarlos es determinístico.

Hay dos métodos que son los más utilizados para la generación. Ambos se basan en la siguiente definición:

**Definición.** Dos enteros $a$ y $b$ son congruentes módulo $m$ si su diferencia es un múltiplo entero de $m$ y se expresa como
$$𝑎 \equiv 𝑏 ( \text{𝑚o𝑑 } 𝑚)$$
Como consecuencia:
- $(a-b)$ es divisible entre $m$
- $a$ y $b$ dan el mismo residuo al ser divididos entre $m$


#### Método congruencial multiplicativo

Generar una secuencia de números pseudoaleatorios uniformes en el intervalo (0, 1) usando la siguiente fórmula recursiva:

$$n_{i+1} = a n_i \mod m$$
- $m$ debe ser tan grande como sea posible, dependiendo de los bits por palabra que maneje la computadora, descontando el bit del signo ($b$). Por lo tanto: $m = 2^b$
- $a$ debe satisfacer que $a \approx 2^{(b+1)/2}$ y que $a \equiv \pm 3 \mod 8$. La segunda expresión equivale a $( a-(\pm3 ) )$ es múltiplo de $8$.
- $n_0$: entero positivo impar menor a $m$
- El periodo será de longitud: $m/4$

```python
# n0 : valor inicial
# a : multiplicador
# m : módulo (2^b)
# n : el número de numeros que quiero generar

def multiplicativo(n0,a,m,n):
    secuencia = [] # lista vacía para guardar los números generados
    ni = n0 # inicio con un valor inicial
    for i in range(n): # repito n veces 
        ni = (a * ni) % m # calculo del siguiente número utilizando la fórmula recurrente
        secuencia.append(ni/m) # guardo el número generado en la lista, normalizado entre 0 y 1
    return secuencia # devuelvo la lista con los números generados
```

```python
m = 2**31 
a = 65539 # a≡ +-3 (mod 8)
n0 =123456789
n = 10000

secuencia_mult = multiplicativo(n0,a,m,n)
print(secuencia_mult)
```

#### Método congruencial mixto

Este método genera números pseudoaleatorios con la fórmula: 
$$n_{i+1} = (a n_i + c) \mod m$$

**Obs.** Se le llama “mixto” porque incluye una constante adicional $c$ (a diferencia del método multiplicativo).

- $m = 2^b$
- $a \approx 2^{(b-1)/2}$, $a \equiv 1 \mod 4$, es decir, $(a-1)$ es múltiplo de 4.
- $c$, $n_0$: enteros positivos impares menores a $m$
- Periodo: tiene longitud $m$

```python
# n0 : valor inicial
# a : multiplicador
# c : constante (impar)
# m : módulo (2^b)
# n : el número de numeros que quiero generar

def mixto(n0,a,c,m,n):
    secuencia = [] # lista vacía para guardar los números generados
    ni = n0 # inicio con un valor inicial
    for i in range(n): # repito n veces 
        ni = (a * ni + c) % m # calculo del siguiente número utilizando la fórmula recurrente
        secuencia.append(ni/m) # guardo el número generado en la lista, normalizado entre 0 y 1
    return secuencia # devuelvo la lista con los números generados
```

```python
m = 2**31
a = 1103515245 # a≡ 1 (mod 4)
c = 12345 # c impar
n0 = 41
n = 10

secuencia_mixto = mixto(n0,a,c,m,n)
print(secuencia_mixto)
```

| **Método**     | **Fórmula**                            | **Ventajas**                                          | **Riesgos / Limitaciones**                                |
|----------------|----------------------------------------|-------------------------------------------------------|------------------------------------------------------------|
| Multiplicativo | \( n_{i+1} = a n_i \mod m \)           | Rápido, menos memoria                                 | Menor aleatoriedad, requiere \( a, m, n_0 \) bien elegidos |
| Mixto          | \( n_{i+1} = (a n_i + c) \mod m \)     | Mejor distribución, periodo completo posible          | Más complejo, pero más robusto                             |
| NumPy (`uniform`) | Motor moderno y validado           | Alta calidad, probado, rápido                         | Caja negra, sin control del generador interno              |


## Metodos de generación de numeros aleatorios no rectangulares

#### Método de la transformada inversa

El método utiliza la función de distribución $F(x)$ de la distribución que se va a simular 
$$F(x) = \int_{-\infty}^{x} f(t)\,dt$$

Como se sabe, los valores de $F(x)$ están en el intervalo $(0,1)$ al igual que los números rectangulares $U$. 

El método genera un $U$ y trata de determinar el valor de la variable aleatoria para la cual $F(x)$ sea igual a $U$.

Si $U \in (0,1)$:
$$ F(x) = U \quad \Rightarrow \quad x = F^{-1}(U)$$

$\mathbf{Teorema}.$ Sea $X$ una variable aleatoria real. Supongamos que su función de distribución $F$ es estrictamente creciente (por lo que $F$ es una biyección de $\mathbb{R}$ sobre $(0,1)$ y podemos denotar por $F^{-1}$ a su inversa). Sea $U\sim \text{unif}[0,1]$. Entonces $F^{-1}(U)$ tiene la misma ley que $X$.

Si $F$ no es estrictamente creciente, hemos visto que el teorema precedente sigue siendo válido bajo la condición de definir 
$$F^{-1}(u)=\inf\{ x\in \mathbb{R} : F(x)\geq u\},$$
la inversa generalizada de $F$.

```python
# Vamos a simular una v.a Bernoulli usando una uniforme
import random
def bernoulli(p):
    U = random.random() # U --> Unif(0,1)
    if U <= p:
        return 1
    else:
        return 0
```

## Ejemplo: Distribución exponencial

Sea $\lambda>0$, entonces a través del teorema se puede generar una v.a. exponencial.

Sea $U \sim Unif[0,1]$, entonces si $X\sim Exp(\lambda)$, entonces

$$f_{X}(x) = \lambda e^{-\lambda x}$$
y que 
$$F_X(x) = 1 - e^{-\lambda x}$$

Sabemos que $1-e^{-\lambda x} = U$, entonces $1-U = e^{-\lambda x}$, y tomando logaritmo, se tiene que 
$ \ln(1-U) = -\lambda x$ y como $U$ es uniforme entonces $U \sim 1-U$, por lo que depejando a $x$, se tiene que 
$$ X = -\frac{\ln(U)}{\lambda} \sim Exp(\lambda) $$


```python
import numpy as np
import plotly.graph_objects as go

np.random.seed(123)

# Parámetros
lambd = 4
n = 10000

# Simulación exponencial
uniformes = np.random.random(n)
exponenciales = -np.log(uniformes)/lambd

# Curva teórica
x_vals = np.linspace(0, exponenciales.max(), 300)
y_vals = lambd * np.exp(-lambd * x_vals)

# Crear figura
fig = go.Figure()
fig.add_trace(go.Histogram(
    x=exponenciales.tolist(),          # convertir a lista
    nbinsx=40,
    histnorm='probability density',
    marker_color='skyblue',
    name='Simulación'
))

fig.add_trace(go.Scatter(
    x=x_vals.tolist(),                 # convertir a lista
    y=y_vals.tolist(),
    mode='lines',
    name='Densidad teórica',
    line={'color':'darkblue'}          # usar diccionario literal
))

fig.update_layout(
    title='Distribución Exponencial simulada vs teórica (λ = 4)',
    xaxis_title='x',
    yaxis_title='Densidad',
    bargap=0.05
)
fig.show()
```

## Ejemplo: Distribución Uniforme Continua
$$f(x) = \frac{1}{b-a}, \quad a \leq x \leq b$$
$$F(x) = \frac{x-a}{b-a} = U \Rightarrow X = a + (b-a)U \sim Unif(a,b)$$

```python
# Parámetros de la uniforme continua
import numpy as np
import plotly.graph_objects as go

a = 2
b = 5
n = 10000 # número de simulaciones

uniformes = np.random.random(n) # Generamos n valores uniformes
uniforme_continua = a + (b - a) * uniformes # Transformada inversa

# Hacemos la densidad teórica
x_vals = np.linspace(a, b, 300)
y_vals = np.ones_like(x_vals) / (b - a)  # Densidad constante entre a y b

# Gráfico
fig = go.Figure()

fig.add_trace(go.Histogram(
    x=uniforme_continua,
    nbinsx=40,
    histnorm='probability density',
    marker_color='orange',
    name='Simulación'
))

fig.add_trace(go.Scatter(
    x=x_vals,
    y=y_vals,
    mode='lines',
    name='Densidad teórica',
    line_color='red'
))

fig.update_layout(
    title='Distribución Uniforme Continua simulada vs teórica Unif[2, 5]',
    xaxis_title='x',
    yaxis_title='Densidad',
    bargap=0.05
)

fig.show()
```

$\mathbf{Ejemplo}$ (Variable aleatoria Cauchy) Simulemos la variable aleatoria de Cauchy de parámetro $1$ que tiene por función de densidad 
$$f_{X}(x) = \frac{1}{\pi}\frac{1}{1+x^{2}}, \quad x\in\mathbb{R}$$

**Ejercicio:** 
- Corroborar que la función así definida es una función de densidad de probabilidad
- Encontrar $F_X$
- Crear una gráfica en Python que muestre su forma
- Generar v.a. Cauchy a partir de una uniforme

Por el teorema anterior, para $u\in (0,1)$
$$u=\frac{1}{\pi}\arctan(x)+\frac{1}{2} \qquad \text{si y sólo si} \qquad x=\tan\left({\pi}\left(u-\frac{1}{2} \right) \right).$$

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# 1. Definimos la densidad
def f_cauchy(x):
    return 1 / (np.pi * (1 + x**2))
```

```python
# 2. Verificamos que integra 1
integral, error = quad(f_cauchy, -np.inf, np.inf)
print("Integral de f(x):", integral)
```

```python
# 3. Función de distribución F(x)
def F_cauchy(x):
    return (1/np.pi) * np.arctan(x) + 0.5
```

```python
# 4. Graficamos densidad y distribución
x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(12,5))

# Densidad
plt.subplot(1,2,1)
plt.plot(x, f_cauchy(x), 'm')
plt.title("Densidad de Cauchy(0,1)")
plt.xlabel("x")
plt.ylabel("f(x)")

# Distribución
plt.subplot(1,2,2)
plt.plot(x, F_cauchy(x), 'c')
plt.title("Función de Distribución Cauchy(0,1)")
plt.xlabel("x")
plt.ylabel("F(x)")

plt.tight_layout()
plt.show()
```

Calcular \( P(X < 1) \) y \( P(-1 < X < 1) \).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import cauchy

# Parámetros
x0 = 0      # ubicación (mediana)
gamma = 1   # escala

#  Cálculamos las  probabilidades
p_menos_1_1 = cauchy.cdf(1, loc=x0, scale=gamma) - cauchy.cdf(-1, loc=x0, scale=gamma)
p_menos_1 = cauchy.cdf(1, loc=x0, scale=gamma)

print(f"P(X < 1) = {p_menos_1:.4f}")
print(f"P(-1 < X < 1) = {p_menos_1_1:.4f}")

# Simulación de datos
np.random.seed(42)
muestras = cauchy.rvs(loc=x0, scale=gamma, size=10000)

#  Limitar valores extremos (colas muy grandes)
muestras_recortadas = muestras[(muestras > -10) & (muestras < 10)]

#  Graficar
x = np.linspace(-10, 10, 500)
plt.figure(figsize=(10,5))
plt.hist(muestras_recortadas, bins=80, density=True, alpha=0.6, color='skyblue', label='Datos simulados')
plt.plot(x, cauchy.pdf(x, loc=x0, scale=gamma), 'r-', lw=2, label='Densidad teórica')
plt.title("Distribución de Cauchy(0, 1)")
plt.xlabel("x")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

# Problema de Aguja de Bufon

```python
# Parámetros del problema

L = 1 # longitud de la aguja
d = 1 # distancia entre las líneas paralelas
num_lanzamientos = 100000 # número total de lanzamientos
step = 1000 # el número de pasos

# Vamos a crear mis listas para guardar los resultados
pi_estimados = [] # almacena los valores de pi estimados
lanzamientos_realizados = [] # almacena el número de lanzamientos realizados

# Contador de los cruces de la línea
# Para saber cuantas agujas cruzan una linea
cruces = 0

# Simulación de los lanzamientos de la aguja y la estimación de pi en pasos 

for i in range(1, num_lanzamientos + 1):
    # Posición del centro de la aguja (uniforme entre 0 y d/2)
    centro = np.random.uniform(0, d/2)

    # Ángulo aleatorio entre 0 y pi (ángulo en radianes)
    angulo = np.random.uniform(0, np.pi)

    #Distancia desde el centro de la aguja hasta la línes más cercana
    distancia = (L / 2) * np.sin(angulo)

    # Verificamos si la aguja cruza una línea
    if centro <= distancia: # si la distancia al centro es menor o igual a la distancia calculada
        cruces += 1 # incrementamos el contador de cruces

    # Estimacion de pi en pasos

    if i % step == 0 and cruces > 0:
        pi_estimado = (2 * L * i) / (d * cruces)
        pi_estimados.append(pi_estimado)
        lanzamientos_realizados.append(i)

print(f"Estimación de pi: {pi_estimados}")

# Graficamos la estimación de pi 
plt.figure(figsize=(10, 6))
plt.plot(lanzamientos_realizados, pi_estimados, label='Estimación de π', color='blue')
plt.axhline(np.pi, color='red', linestyle='--', label='Valor real de π')
plt.xlabel('Número de lanzamientos')
plt.ylabel('Estimación de π')
plt.title('Estimación de π usando el Problema de la Aguja de Buffon')
plt.legend()
plt.show()
```
